Aging Bucket Analysis

```python
# Sum amounts for each aging category
aging_buckets = FinanceData[['l30_amount', 'l60_amount', 'l90_amount', 'l120_amount', 'l150_amount', 'l180_amount']].sum()

# Plot the aging bucket amounts
aging_buckets.plot(kind='bar', title='Aging Analysis')
plt.xlabel='Aging Categories'
plt.ylabel='Amount'
plt.show()
```

Financial Performance by Item


```python
total_amount_item= FinanceData.groupby('item')['total_amount'].sum()
total_amount_item
```

```python
total_amount_item.plot(kind='bar', title='Total Amount by Item')
plt.xlabel='Item'
plt.ylabel='Total Amount'
plt.show()
```

Trends Over Time


```python
total_amount_period = FinanceData.groupby('period_code')['total_amount'].sum()
total_amount_period
```

```python
total_amount_period.plot(kind='line', title='Financial Amount Trend Over Time')
plt.xlabel='Period Code'
plt.ylabel='Total Amount'
plt.show()
```

Bad Debt and Aging Ratio Analysis

```python
FinanceData['bad_amount_ratio'] = FinanceData['bad_amount']/FinanceData['total_amount']
FinanceData['aging_sum'] = FinanceData[['l30_amount', 'l60_amount', 'l90_amount', 'l120_amount']].sum(axis=1)
FinanceData['aging_ratio'] = FinanceData['aging_sum'] / FinanceData['total_amount']
print(FinanceData[['period_code', 'bad_amount_ratio', 'aging_ratio']])
```

Correlation Heatmap

The image below is a correlation heatmap that visualizes the relationships between different financial amounts in the dataset, such as bad_amount, badi_amount, g1_amount, and others. Each cell in the heatmap represents the correlation between two financial variables, with colors indicating the strength and direction of the correlation. Warm colors (reds) show positive correlations, while cool colors (blues) show negative correlations. The values inside the cells provide exact correlation coefficients, helping to identify how strongly these amounts are related to each other. This heatmap is useful for identifying patterns, trends, and potential dependencies between different financial metrics.

```python
corr = FinanceData[['bad_amount', 'badi_amount', 'g1_amount', 'l1_amount', 
           'l120_amount', 'l150_amount', 'l180_amount', 'l30_amount',
           'l60_amount', 'l90_amount', 'total_amount']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Financial Amounts')
plt.show()
```
